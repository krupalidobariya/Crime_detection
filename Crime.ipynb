# Data Visualisation and Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Dataset=pd.read_csv('crime.csv')
Dataset
```

```python
# Removing all null values

Dataset.dropna(how='any',axis=0,inplace=True)
Dataset
```

```python
# Removing coloumn which is not necessary.
Dataset.drop(['HUNDRED_BLOCK','X','Y'],axis=1,inplace=True)
```

```python
# Counting numbers of the particular crime.
crimes_num=Dataset.TYPE.value_counts().iloc[:10]
crimes_num
```

```python
# Plotting the crimes according to the number
Dataset.TYPE.value_counts().iloc[:10].sort_values().plot(kind='barh')
plt.xlabel('Number Of Crimes')
plt.ylabel('Type Of Crime')
plt.title('Different Crimes according to the Number')
plt.show()
```

```python
Dataset.sort_index(axis=0,inplace=True)
```

```python
# 
X=Dataset.iloc[:,[1,2,3,4,5,6,7,8]].values
Y=Dataset.iloc[:,0].values
```

```python
# Here we are taking care of categorical data for X
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_x=LabelEncoder()
X[:,5]=labelencoder_x.fit_transform(X[:,5])
onehotencoder=OneHotEncoder(categorical_features=[5])
X=onehotencoder.fit_transform(X).toarray()


```

```python
# Similarly for taking care of categorical data for y
labelencoder_x=LabelEncoder()
Y=labelencoder_x.fit_transform(Y)
```

```python
# Here we are avoiding dummy variable trap
X=X[:,1:]
X=pd.DataFrame(X)
X.head()
```

```python
#Splitting the data into train set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
X_train=pd.DataFrame(X_train)
X_train.head()
```

```python
#Feature Scaling of latitude and longitude
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train[[23,28,29]]=sc_X.fit_transform(X_train[[23,28,29]])
X_test[[23,28,29]]=sc_X.transform(X_test[[23,28,29]])
```

```python

from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=95,criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
# Predicting the test set results..
Y_pred=classifier.predict(X_test)
```

```python
# Checking the model accuracy...
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
# Making the confusion matrix... So that we can know how many wrong predictions
# our model has make....
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
```
